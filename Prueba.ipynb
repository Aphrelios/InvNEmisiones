```python
# Leer el csv de inventarios de emisiones por año
# Arrojaba un error de codificacion por lo que agregue
# encoding_errors = "ignore"
year=1990
Inv90 = pd.read_csv(f"\\Users\\ALFREDO\\Documents\\Inventario\\{year}\\{year}.csv", encoding = "UTF-8", encoding_errors = "ignore", thousands=",")
Inv90 = Inv90.rename(columns={"Ao": "Año", "Categoria IPCC": "Categoría IPCC", "Sub Categora IPCC": "Sub Categoría IPCC"})
Inv90["Sector IPCC"] = Inv90["Sector IPCC"].apply(lambda x: x.replace("Energa", "Energía"))
Inv90
```

```python
# Fijo las columnas que no quiero que se apilen
fijas = Inv90.columns[:6]
fijas
```

```python
# Aquí incluyó las columnas que me interesa apilar
variables = Inv90.columns[6:]
variables
```

```python
# Con esta línea apilo creo un arreglo con las columnas fijadas y las columnas apiladas por gas y emisión
stack_columns = Inv90.set_index(list(fijas)).stack().reset_index()
stack_columns
```

```python
# Renombro las columnas creadas
# No es posible editar las columnas "0" que deberías corresponder a las "Emisiones"
rename = stack_columns.rename(columns = {"level_6": "Gas", 0: "Emision" })
rename
```

```python
# Agrego la columna correspondiente a las unidades
rename["Unidades"] = "Gg de CO2e"
rename
```

```python
rename[["Año", "Categoría IPCC", "Emision"]].groupby(["Año", "Categoría IPCC"]).sum().reset_index()
```
